```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Salary.csv')
df.head()
```

```python
reg = LinearRegression()
x = df[['age','experience']]
y = df['income']
reg.fit(x,y)
```

```python
print('COEF is:',reg.coef_)
print('Intercept is:',reg.intercept_)
```

```python
# Example prediction
prediction = reg.predict([[50, 30]])
predicted_income = prediction[0]
print("Predicted Income:", predicted_income)

# Create a scatter plot for age and experience against income
plt.figure(figsize=(8, 6))

# Plotting the actual data points
plt.scatter(df['age'], df['income'], color='blue', label='Actual Income')
plt.scatter(df['experience'], df['income'], color='green', label='Actual Income')

# Adding the predicted income point
plt.scatter(35, predicted_income, color='red', marker='o', s=100, label='Predicted Income')

# Adding labels and title
plt.xlabel('Age and Experience')
plt.ylabel('Income')
plt.title('Actual and Predicted Income vs. Age and Experience')
plt.legend()
plt.show()
```
